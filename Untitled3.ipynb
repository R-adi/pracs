```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from tensorflow.keras.datasets import fashion_mnist

(train_x, train_y), (test_x, test_y) = fashion_mnist.load_data()

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, MaxPooling2D, Conv2D

model = Sequential()

model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))

# Adding maxpooling layer to get max value within a matrix
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(128, activation="relu"))
model.add(Dense(10, activation="softmax"))

model.summary()

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(train_x.astype(np.float32), train_y.astype(np.float32), epochs=2, validation_split=0.2)

loss, acc = model.evaluate(test_x, test_y)

labels = ['t_shirt', 'trouser', 'pullover', 'dress', 'coat', 'sandal', 'shirt', 'sneaker', 'bag', 'ankle_boots']

# Predicting on 10 images (test_x[1:11])
predictions = model.predict(test_x[1:11])

# Display each predicted label and corresponding image
for i in range(10):
    label = labels[np.argmax(predictions[i])]  # Find the predicted label for each image
    print(f"Image {i+1}: {label}")
    
    # Plot the image
    plt.imshow(test_x[i+1], cmap='gray')  # Display image with index i+1
    plt.title(label)  # Set the title as the predicted label
    plt.show()
```
